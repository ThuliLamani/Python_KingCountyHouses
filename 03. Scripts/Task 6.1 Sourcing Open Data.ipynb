The max for 'bedrooms' equals 33, which is irregular . Changing that to 3, which is closer to the mean. 


```python
# Replace '33' with '3'
df_kc_house.loc[ df_kc_house['bedrooms'] == 33, 'bedrooms'] = 3
```

```python
#Check 'bedrooms' values
df_kc_house['bedrooms'].value_counts()
```

The minimum value for 'bedrooms' was 0, and there are 13 instances where this is true. However, I will not remove or impute these values. 

The minimum for 'bathrooms' is 0, which makes sense as a bathroom is a standard requirement for a home.

```python
#Check how many instances where the value for 'bathrooms' equals 0. 
df_kc_house['bathrooms'].value_counts()
```

```python
#Replace variables in 'bathrooms' where value is 0 to mean (2) 
df_kc_house.loc[ df_kc_house['bathrooms'] == 0, 'bathrooms'] = 2
```
